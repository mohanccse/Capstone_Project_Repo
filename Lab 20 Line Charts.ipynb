### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python

# Keep relevant columns
df_age = df[['Age', 'ConvertedCompYearly']].dropna()

# Group by Age category & compute median compensation
age_group_median = df_age.groupby('Age')['ConvertedCompYearly'].median().sort_index()

# Plot line chart
plt.figure(figsize=(10,5))
plt.plot(age_group_median.index, age_group_median.values, marker='o')
plt.xticks(rotation=45)
plt.title("Median Yearly Compensation by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Median ConvertedCompYearly")
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
import re

# Function to convert age group to numeric midpoint
def age_to_midpoint(age_str):
    match = re.findall(r'\d+', str(age_str))
    if len(match) == 2:   # e.g., "25-34 years old"
        return (int(match[0]) + int(match[1])) / 2
    else:
        return None  # For values like "Under 18 years old"

df['Age_mid'] = df['Age'].apply(age_to_midpoint)

# Filter ages between 25 and 45
df_mid = df[(df['Age_mid'] >= 25) & (df['Age_mid'] <= 45)]

# Compute median compensation within these numeric age groups
age_mid_median = df_mid.groupby('Age_mid')['ConvertedCompYearly'].median().sort_index()

# Plot
plt.figure(figsize=(10,5))
plt.plot(age_mid_median.index, age_mid_median.values, marker='o')
plt.title("Median Yearly Compensation for Ages 25 to 45")
plt.xlabel("Age (Midpoint)")
plt.ylabel("Median ConvertedCompYearly")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Convert WorkExp to numeric (coerce errors → NaN)
df['WorkExp_num'] = pd.to_numeric(df['WorkExp'], errors='coerce')

# Keep only needed columns
df_exp = df[['WorkExp_num', 'JobSatPoints_6']].dropna()

# Compute average (or median) job satisfaction by experience
job_sat_by_exp = df_exp.groupby('WorkExp_num')['JobSatPoints_6'].mean()

# Plot
plt.figure(figsize=(10,5))
plt.plot(job_sat_by_exp.index, job_sat_by_exp.values, marker='o')
plt.title("Job Satisfaction (JobSatPoints_6) by Experience Level")
plt.xlabel("Years of Experience (WorkExp)")
plt.ylabel("Average Job Satisfaction")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
# Use same numeric conversion
df['WorkExp_num'] = pd.to_numeric(df['WorkExp'], errors='coerce')

df_comp = df[['WorkExp_num', 'ConvertedCompYearly']].dropna()

# Median compensation by experience
comp_by_exp = df_comp.groupby('WorkExp_num')['ConvertedCompYearly'].median()

plt.figure(figsize=(10,5))
plt.plot(comp_by_exp.index, comp_by_exp.values, marker='o')
plt.title("Median Yearly Compensation by Experience Level")
plt.xlabel("Years of Experience (WorkExp)")
plt.ylabel("Median ConvertedCompYearly")
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
df_js = df[['WorkExp_num', 'JobSatPoints_6']].dropna()

job_sat_trend = df_js.groupby('WorkExp_num')['JobSatPoints_6'].mean()

plt.figure(figsize=(10,5))
plt.plot(job_sat_trend.index, job_sat_trend.values, marker='o')
plt.title("Job Satisfaction (JobSatPoints_6) Across Experience Levels")
plt.xlabel("Years of Experience (WorkExp)")
plt.ylabel("Average Job Satisfaction")
plt.grid(True)
plt.tight_layout()
plt.show()
```
