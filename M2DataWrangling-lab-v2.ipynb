<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
# 5.2 Impute missing values in numerical columns with mean or median
if 'ConvertedCompYearly' in df.columns:
    median_value = df['ConvertedCompYearly'].median()
    df['ConvertedCompYearly'].fillna(median_value, inplace=True)
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python

# 5.3 Impute missing values in categorical columns with most frequent value
if 'RemoteWork' in df.columns:
    most_frequent = df['RemoteWork'].mode()[0]
    df['RemoteWork'].fillna(most_frequent, inplace=True)
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
# 6.1 Apply Min-Max Scaling to normalize the ConvertedCompYearly column
scaler = MinMaxScaler()
if 'ConvertedCompYearly' in df.columns:
    df['ConvertedCompYearly_Scaled'] = scaler.fit_transform(df[['ConvertedCompYearly']])
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
# 6.2 Log-transform the ConvertedCompYearly column to reduce skewness
# Adding 1 to avoid log(0)
if 'ConvertedCompYearly' in df.columns:
    df['ConvertedCompYearly_Log'] = np.log1p(df['ConvertedCompYearly'])
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python

# 7. Feature Engineering
# 7.1 Create a new column ExperienceLevel based on YearsCodePro column
def experience_level(years):
    if pd.isna(years):
        return 'Unknown'
    try:
        years = float(years)
    except:
        return 'Unknown'
    if years < 2:
        return 'Junior'
    elif 2 <= years < 5:
        return 'Intermediate'
    elif 5 <= years < 10:
        return 'Senior'
    else:
        return 'Expert'

if 'YearsCodePro' in df.columns:
    df['ExperienceLevel'] = df['YearsCodePro'].apply(experience_level)

# Example print to check the transformations
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Scaled', 'ConvertedCompYearly_Log', 'YearsCodePro', 'ExperienceLevel']].head())
```
